```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.pyplot import figure
mpl.rc('font',family='Serif')
import re
np.set_printoptions(precision=2)
```

```python
def getData(file):
    data = pd.read_csv(file)
    data_array = np.array(data)
    wavelength = data_array[:,0]
    intensity = data_array[:,1]
    
    return wavelength, intensity
```

```python
lam_H, I_H = getData('hydrogen.csv')
lam_He, I_He = getData('helium.csv')
lam_N, I_N = getData('nitrogen.csv')
lam_K, I_K = getData('K.csv')
lam_I, I_I = getData('I.csv')
lam_Hg_control, I_Hg_control = getData('Hg_normal.csv')
lam_Hg_clear, I_Hg_clear = getData('Hg_clear.csv')
lam_Hg_sunglasses, I_Hg_sunglasses = getData('Hg_sunglasses.csv')
lam_CFL, I_CFL = getData('CFL.csv')
lam_halogen, I_halogen = getData('halogen.csv')
```

```python
figure(figsize=(8,6), dpi=80)
plt.plot(lam_H, I_H)
plt.grid()
plt.xlabel('$\lambda$ (nm)', fontsize=14)
plt.ylabel('Relative intensity (counts)', fontsize=14)
plt.tick_params(labelsize=12)
plt.savefig('H_emission.png')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5), sharex=True, sharey=True)
ax[0].plot(lam_He, I_He, 'darkorange', label="Helium Emission")
ax[1].plot(lam_N, I_N, label='Nitrogen Emission')
for i in range(2):
    ax[i].grid()
    ax[i].legend(fontsize=14)
    ax[i].tick_params(labelsize=12)
ax[0].set_ylabel('Relative Intensity (counts)', fontsize=14)
ax[0].set_xlabel('$\lambda$ (nm)', fontsize=14)
ax[1].set_xlabel('$\lambda$ (nm)', fontsize=14)
fig.tight_layout()
fig.savefig('HevsN.png')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5), sharex=True, sharey=True)
ax[0].plot(lam_K, I_K, 'darkorange', label="Mystery Tube K Emission")
ax[1].plot(lam_I, I_I, label='Mystery Tube I Emission')
for i in range(2):
    ax[i].grid()
    ax[i].legend(fontsize=12)
    ax[i].tick_params(labelsize=12)
ax[0].set_ylabel('Relative Intensity (counts)', fontsize=14)
ax[0].set_xlabel('$\lambda$ (nm)', fontsize=14)
ax[1].set_xlabel('$\lambda$ (nm)', fontsize=14)
fig.tight_layout()
fig.savefig('KvsI.png')
```

```python
figure(figsize=(8,6), dpi=80)
plt.plot(lam_CFL, I_CFL, label="CFL Bulb Emission")
plt.grid()
plt.legend()
plt.xlabel('$\lambda$ (nm)', fontsize=14)
plt.ylabel('Relative intensity (counts)', fontsize=14)
plt.tick_params(labelsize=12)
plt.savefig('CFL_emission.png')

figure(figsize=(8,6), dpi=80)
plt.plot(lam_halogen, I_halogen, 'r', label="Halogen Bulb Emission")
plt.grid()
plt.xlabel('$\lambda$ (nm)', fontsize=14)
plt.ylabel('Relative intensity (counts)', fontsize=14)
plt.tick_params(labelsize=12)
plt.savefig('halogen_emission.png')
```

```python
fig, ax = plt.subplots(3, 1, figsize=(9, 10), sharex=True, sharey=True)
ax[0].plot(lam_Hg_control, I_Hg_control, 'r', label="Mercury Control")
ax[1].plot(lam_Hg_sunglasses, I_Hg_sunglasses, 'darkorange', label='Mercury with Sunglasses')
ax[2].plot(lam_Hg_clear, I_Hg_clear, label='Mercury with Clear Safety Glasses')
for i in range(3):
    ax[i].grid()
    ax[i].tick_params(labelsize=12)
    ax[i].legend(fontsize=12)
ax[1].set_ylabel('Relative Intensity (counts)', fontsize=14)
ax[2].set_xlabel('$\lambda$ (nm)', fontsize=14)
fig.tight_layout()
fig.savefig('Hg.png')
```
